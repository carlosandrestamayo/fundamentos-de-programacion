# ğŸ”€ La estructura condicional `if...else` en Python

Si la condiciÃ³n es **verdadera**, se ejecuta el bloque del `if`;

## ğŸ§© Estructura bÃ¡sica

## ğŸ”‘ Partes fundamentales de una estructura `if...else`

- Puedes usar varias estructuras `if...else` en tu programa para evaluar distintos casos.  
- La **sangrÃ­a** en Python es **obligatoria** para delimitar los bloques de cÃ³digo.


# ğŸ” La estructura condicional `if...elif...else` en Python

La estructura condicional `if...elif...else` permite **evaluar mÃºltiples condiciones** de forma secuencial.  
Se utiliza cuando **existen mÃ¡s de dos posibles caminos** para la ejecuciÃ³n del programa.

El programa evalÃºa cada condiciÃ³n en orden:  
- Si una condiciÃ³n se cumple (`True`), se ejecuta su bloque y se ignoran las demÃ¡s.  
- Si ninguna se cumple, se ejecuta el bloque del `else`.

---

## ğŸ§© Estructura general

```python
if condiciÃ³n1:
    # bloque si condiciÃ³n1 es verdadera
elif condiciÃ³n2:
    # bloque si condiciÃ³n2 es verdadera
elif condiciÃ³n3:
    # bloque si condiciÃ³n3 es verdadera
else:
    # bloque si ninguna condiciÃ³n es verdadera
```

---

## ğŸ”‘ Partes fundamentales de una estructura `if...elif...else`

1. **Condiciones lÃ³gicas**  
   Cada `if` o `elif` evalÃºa una expresiÃ³n booleana (`True` o `False`).

2. **Bloques de cÃ³digo**  
   Cada condiciÃ³n tiene su propio bloque indentado.

3. **Orden de evaluaciÃ³n**  
   Python evalÃºa de arriba hacia abajo y se detiene en la primera condiciÃ³n verdadera.

4. **Bloque `else` (opcional)**  
   Se ejecuta solo si **ninguna condiciÃ³n anterior** es verdadera.

---

## ğŸ’¡ Ejemplo prÃ¡ctico

```python
temperatura = 30

if temperatura < 10:
    print("Hace frÃ­o")
elif temperatura < 25:
    print("El clima es agradable")
else:
    print("Hace calor")
```

---

### ğŸ“˜ Salida

```
Hace calor
```

---

## âš ï¸ Importante

- Puedes tener **varios bloques `elif`**, pero solo **un `if`** y **un `else`**.  
- El bloque `else` es opcional.  
- La ejecuciÃ³n se detiene **en la primera condiciÃ³n verdadera**.  
- Todas las condiciones deben estar **correctamente indentadas**.

---

## ğŸ§  En resumen

| Estructura | DescripciÃ³n | CuÃ¡ndo usar |
|-------------|--------------|--------------|
| `if` | EvalÃºa una condiciÃ³n | Cuando hay una sola opciÃ³n |
| `if...else` | EvalÃºa dos caminos | Cuando hay dos posibles resultados |
| `if...elif...else` | EvalÃºa mÃºltiples condiciones | Cuando hay mÃ¡s de dos opciones |

## ğŸ”¹ 1. La funciÃ³n `len()`

- La funciÃ³n **`len()`** devuelve la **longitud** (cantidad de elementos) de una cadena, lista, tupla u otro objeto iterable.  
- Es muy usada para recorrer textos o listas con ciclos como `while` o `for`.

## ğŸ”¹ 2. El operador `in`

El operador **`in`** se utiliza para **verificar si un elemento estÃ¡ dentro de una cadena, lista o conjunto**.  

ğŸ‘‰ Devuelve un valor booleano:

### ğŸ“Œ Ejemplos
