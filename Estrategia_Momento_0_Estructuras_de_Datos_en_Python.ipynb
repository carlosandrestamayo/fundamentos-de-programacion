# üì¶ Introducci√≥n a las Estructuras de Datos en Python

En programaci√≥n, una **estructura de datos** es una forma organizada de **guardar, gestionar y acceder a la informaci√≥n**.  
Nos permiten trabajar con colecciones de valores de manera eficiente, adapt√°ndose a distintas necesidades.

En Python, las principales estructuras de datos **b√°sicas** son:

- **Listas (`list`)** ‚Üí Colecciones **ordenadas y mutables**.  
- **Tuplas (`tuple`)** ‚Üí Colecciones **ordenadas e inmutables**.  
- **Conjuntos (`set`)** ‚Üí Colecciones **desordenadas y sin duplicados**.  
- **Diccionarios (`dict`)** ‚Üí Colecciones de pares **clave ‚Üí valor**.  

---

## üîë ¬øPor qu√© son importantes?
- Permiten almacenar varios datos en una sola variable.  
- Facilitan operaciones como agregar, buscar, modificar o eliminar elementos.  
- Son la base para resolver problemas complejos de programaci√≥n.  

---

## üß© Analog√≠as cotidianas
- **Lista** ‚Üí Una lista de compras (ordenada y modificable).  
- **Tupla** ‚Üí Una fecha de nacimiento (datos fijos que no cambian).  
- **Conjunto** ‚Üí Un grupo de amigos (sin repetidos, el orden no importa).  
- **Diccionario** ‚Üí Una agenda telef√≥nica (nombre ‚Üí n√∫mero).  

---

# üìù Listas (`list`) en Python

Imagina una **lista de compras** en el supermercado:  
- Tiene un **orden** (primero leche, luego pan, despu√©s huevos).  
- Puedes **agregar** cosas nuevas.  
- Puedes **quitar** lo que ya no necesitas.  
- Puedes **modificar** lo que apuntaste mal.  

En Python, esa idea se representa con las **listas** (`list`).

---

## üì¶ ¬øQu√© es una lista?
- Una **colecci√≥n ordenada** de elementos.  
- Es **mutable** ‚Üí se puede cambiar (agregar, quitar o modificar elementos).  
- Se define con corchetes `[]`.  
- Puede contener **n√∫meros, textos, booleanos, o incluso otras listas**.

---

## ‚ú® Creaci√≥n de listas
```python
# Lista de frutas
frutas = ["manzana", "pera", "mango"]

# Lista de n√∫meros
numeros = [1, 2, 3, 4, 5]

# Lista con datos mixtos
mixta = [10, "hola", True, 3.14]


## üîë Indexaci√≥n en listas
Los elementos de una lista se acceden por su **posici√≥n (√≠ndice)**:  
- El primer elemento est√° en la posici√≥n **0**.  
- El √∫ltimo elemento se puede acceder con un √≠ndice negativo **-1**.  

Tambi√©n se pueden obtener **sublistas** usando *slicing*.
```
---

```python
frutas = ["manzana", "pera", "mango", "uva"]

print(frutas[0])    # Primer elemento
print(frutas[2])    # Tercer elemento
print(frutas[-1])   # √öltimo elemento

print(frutas[1:3])  # Sublista ['pera', 'mango']
print(frutas[:2])   # Sublista desde inicio ['manzana', 'pera']
print(frutas[2:])   # Sublista desde √≠ndice 2 ['mango', 'uva']

```

---

## üõ†Ô∏è M√©todos principales de listas
Las listas tienen varios m√©todos que permiten **agregar, eliminar o modificar** elementos.


```python

frutas = ["manzana", "pera", "mango"]

frutas.append("uva")        # Agregar al final
frutas.insert(1, "kiwi")    # Insertar en posici√≥n espec√≠fica
frutas.remove("pera")       # Eliminar por valor
frutas.pop()                # Eliminar y devolver el √∫ltimo

print(frutas)
print("Cantidad de elementos:", len(frutas))

```
---

## üìè Dimensiones de las listas
Las listas pueden ser:
1. **Unidimensionales** ‚Üí Una sola fila de datos.  
2. **Bidimensionales (listas de listas)** ‚Üí Como una tabla o matriz.  
3. **Multidimensionales** ‚Üí Listas dentro de listas, de m√°s niveles.

```python

# Lista unidimensional
numeros = [1, 2, 3, 4, 5]
print(numeros[2])  # Accede al n√∫mero 3

# Lista bidimensional (matriz 3x3)
matriz = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

print(matriz[0][1])  # Accede al n√∫mero 2
print(matriz[2][0])  # Accede al n√∫mero 7

# Lista tridimensional (ejemplo simple)
cubos = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(cubos[1][0][1])  # Accede al n√∫mero 6

```

# üîí Tuplas (`tuple`) en Python

Imagina tu **fecha de nacimiento**:  
- D√≠a, mes y a√±o siempre ser√°n los mismos.  
- No necesitas cambiarlos nunca.  

En Python, cuando queremos guardar datos que **no se deben modificar**, usamos **tuplas** (`tuple`).

---

## üì¶ ¬øQu√© es una tupla?
- Una **colecci√≥n ordenada** de elementos.  
- Es **inmutable** ‚Üí una vez creada, no se puede cambiar.  
- Se define con par√©ntesis `()`.  
- Puede contener **n√∫meros, textos, booleanos u otras tuplas**.  

---

## ‚ú® Creaci√≥n de tuplas
```python
coordenada = (10, 20)
persona = ("Ana", 28, "Ingeniera")
mixta = (3.14, "Python", False)
```

---

## üîë Indexaci√≥n y slicing
```python
frutas = ("manzana", "pera", "mango")

print(frutas[0])    # Primer elemento
print(frutas[2])    # Tercer elemento
print(frutas[-1])   # √öltimo elemento
print(frutas[0:2])  # Subtupla ('manzana', 'pera')
```

---

## üõ†Ô∏è M√©todos de las tuplas
```python
numeros = (1, 2, 2, 3, 4, 2)

print(numeros.count(2))   # 3
print(numeros.index(3))   # 3
```

---

## üìè Dimensiones
```python
punto3D = ((1, 2), (3, 4), (5, 6))

print(punto3D[0])      # (1, 2)
print(punto3D[0][1])   # 2
```

---

## ‚úÖ Resumen
- Una tupla es **ordenada e inmutable**.  
- Se define con `()`.  
- M√©todos principales: `count()` e `index()`.  
- √ötil para representar datos que **no cambian**.  


# üîÄ Conjuntos (`set`) en Python

Imagina un **grupo de amigos**:  
- No importa el **orden** en el que los digas.  
- **No puedes repetir** la misma persona dos veces.  

En Python, esa idea se representa con los **conjuntos** (`set`).

---

## üì¶ ¬øQu√© es un conjunto?
- Una **colecci√≥n desordenada** de elementos.  
- **No admite duplicados**.  
- Se define con llaves `{}` o con la funci√≥n `set()`.  
- √ötil para eliminar repetidos y realizar operaciones de **teor√≠a de conjuntos**.

---

## ‚ú® Creaci√≥n de conjuntos
```python
# Creaci√≥n b√°sica
numeros = {1, 2, 3, 4, 5}

# Elimina duplicados autom√°ticamente
repetidos = {1, 2, 2, 3, 4, 4, 5}
print(repetidos)  # {1, 2, 3, 4, 5}

# Convertir lista en conjunto
lista = [1, 2, 2, 3, 4, 4, 5]
conjunto = set(lista)
print(conjunto)
```

---

## üîë Propiedades importantes
- Los conjuntos son **desordenados** ‚Üí no se accede con √≠ndices.  
- Si necesitas acceder por posici√≥n, primero convi√©rtelo en lista.  

```python
conjunto = {"a", "b", "c"}
print("a" in conjunto)  # True (verifica pertenencia)

lista = list(conjunto)
print(lista[0])  # Convertido a lista para acceder por √≠ndice
```

---

## üõ†Ô∏è M√©todos principales de conjuntos
```python
cursos = {"Python", "C#", "Java"}

cursos.add("Go")        # Agregar
cursos.discard("Java")  # Eliminar sin error si no existe
print(len(cursos))      # Cantidad de elementos
```

---

## üîó Operaciones de conjuntos
Los conjuntos permiten operaciones matem√°ticas cl√°sicas:  

- **Uni√≥n (`|`)** ‚Üí elementos de ambos conjuntos.  
- **Intersecci√≥n (`&`)** ‚Üí elementos en com√∫n.  
- **Diferencia (`-`)** ‚Üí elementos de un conjunto que no est√°n en el otro.  
- **Diferencia sim√©trica (`^`)** ‚Üí elementos que est√°n en uno u otro, pero no en ambos.  

```python
a = {1, 2, 3}
b = {3, 4, 5}

print("Uni√≥n:", a | b)            # {1, 2, 3, 4, 5}
print("Intersecci√≥n:", a & b)     # {3}
print("Diferencia:", a - b)       # {1, 2}
print("Dif. sim√©trica:", a ^ b)   # {1, 2, 4, 5}
```

---

## üìè Conjuntos anidados
Un conjunto no puede contener **otros conjuntos** (porque no son inmutables), pero s√≠ puede contener **tuplas** (que s√≠ son inmutables).

```python
conjunto = {(1, 2), (3, 4)}
print(conjunto)
```

---

## ‚úÖ Resumen
- Un conjunto es **desordenado y sin duplicados**.  
- Se define con `{}` o `set()`.  
- M√©todos √∫tiles: `add()`, `discard()`, `len()`.  
- Operaciones: **uni√≥n, intersecci√≥n, diferencia, diferencia sim√©trica**.  
- √ötil para trabajar con **datos √∫nicos** y operaciones de conjuntos.  


# üìñ Diccionarios (`dict`) en Python

Imagina una **agenda telef√≥nica**:  
- Cada persona tiene un **nombre** (clave).  
- Cada nombre tiene un **n√∫mero** asociado (valor).  

En Python, esa idea se representa con los **diccionarios** (`dict`).

---

## üì¶ ¬øQu√© es un diccionario?
- Una colecci√≥n de pares **clave ‚Üí valor**.  
- Las **claves** deben ser √∫nicas e inmutables (texto, n√∫mero o tupla).  
- Los **valores** pueden ser de cualquier tipo.  
- Se define con llaves `{}`.  

---

## ‚ú® Creaci√≥n de diccionarios
```python
# Diccionario de un estudiante
estudiante = {
    "nombre": "Ana",
    "edad": 20,
    "carrera": "Ingenier√≠a"
}

print(estudiante)
```

---

## üîë Acceso a los valores
```python
print(estudiante["nombre"])      # Acceso directo a valor
print(estudiante.get("edad"))    # Acceso seguro con .get()
```

Si accedes a una clave inexistente con `[]`, genera error.  
Con `.get()` devuelve `None` (o un valor por defecto si se lo indicas).

---

## üõ†Ô∏è M√©todos principales
```python
estudiante["edad"] = 21              # Modificar valor
estudiante["universidad"] = "UNAL"   # Agregar nuevo par
del estudiante["carrera"]            # Eliminar clave-valor

print(estudiante.keys())    # Todas las claves
print(estudiante.values())  # Todos los valores
print(estudiante.items())   # Todos los pares (clave, valor)
```

---

## üìè Diccionarios anidados
Un diccionario puede contener **otros diccionarios o listas** como valores.

```python
agenda = {
    "Ana": {"tel": "123", "edad": 20},
    "Luis": {"tel": "456", "edad": 25},
    "Marta": {"tel": "789", "edad": 30}
}

print(agenda["Ana"]["tel"])   # 123
print(agenda["Marta"]["edad"]) # 30
```

---

## ‚úÖ Resumen
- Un diccionario es una colecci√≥n de pares **clave ‚Üí valor**.  
- Se define con `{}`.  
- Las **claves** son √∫nicas e inmutables.  
- M√©todos √∫tiles: `.keys()`, `.values()`, `.items()`.  
- Permite estructuras m√°s complejas con diccionarios anidados.
