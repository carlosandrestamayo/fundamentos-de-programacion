# üîÑ El ciclo `while` en Python

El ciclo `while` es una **estructura repetitiva** que permite ejecutar un bloque de instrucciones **mientras se cumpla una condici√≥n l√≥gica**.  
Es decir, **repite una acci√≥n hasta que la condici√≥n deje de ser verdadera**.

---

## üß© Estructura b√°sica

```python
while condici√≥n:
    # instrucciones que se repiten
```

## üîë Partes fundamentales del ciclo `while`

El ciclo `while` se compone de tres partes fundamentales:

1. **Inicializaci√≥n**  
   Se define una variable de control antes de iniciar el ciclo.  

2. **Condici√≥n**  
   Se eval√∫a antes de cada repetici√≥n.  
   - Si es verdadera (`True`), el ciclo se ejecuta.  
   - Si es falsa (`False`), el ciclo termina.  

3. **Actualizaci√≥n**  
   Dentro del ciclo se debe modificar la variable de control para que en alg√∫n momento la condici√≥n sea falsa y el ciclo finalice.  


# üìù Ejercicio: Contar del 1 al 10 con un ciclo `while`

## Enunciado
Escribe un programa en Python que muestre en pantalla los n√∫meros del 1 al 10, utilizando un ciclo `while`.  
El programa debe iniciar con la variable en 1 e ir increment√°ndola de uno en uno hasta llegar al 10.  

Este ejercicio sirve para comprender:
- C√≥mo inicializar una variable de control.
- C√≥mo establecer una condici√≥n de repetici√≥n.
- C√≥mo actualizar la variable dentro del ciclo.

---


````python
# Ejemplo: Contar del 1 al 10 usando un ciclo while

# Paso 1: Inicializamos la variable de control
numero = 1

# Paso 2: Establecemos la condici√≥n para que el ciclo se repita
# Mientras el n√∫mero sea menor o igual a 10, el ciclo se ejecutar√°
while numero <= 10:
    # Paso 3: Mostramos el valor actual de la variable
    print("El n√∫mero es:", numero)

    # Paso 4: Actualizamos la variable sumando 1
    numero = numero + 1

# Cuando la condici√≥n deja de cumplirse (numero > 10), el ciclo termina
print("Fin del conteo ‚úÖ")
````

# üìò Acumulador y Contador en Programaci√≥n

Cuando trabajamos con ciclos en programaci√≥n, a menudo necesitamos **guardar informaci√≥n** mientras el ciclo avanza.  
Para eso utilizamos dos conceptos muy importantes: **el contador** y **el acumulador**.  

---

## üî¢ 1. Contador
- Un **contador** es una variable que se utiliza para **contar cu√°ntas veces ocurre algo**.  
- Generalmente se **inicializa en 0 o 1** y se va **incrementando de 1 en 1** en cada repetici√≥n del ciclo.  
- Su funci√≥n principal es **llevar la cuenta** de las iteraciones o elementos procesados.  

üëâ **Ejemplo en Python:**

```python
# Contador: contar hasta 5

contador = 0   # inicializamos en 0

while contador < 5:   # el ciclo se repetir√° mientras contador sea menor que 5
    print("Vuelta n√∫mero:", contador + 1)  # mostramos el n√∫mero de vuelta
    contador = contador + 1   # incrementamos el contador en 1
```


## ‚ûï 2. Acumulador

Un **acumulador** es una variable que se utiliza para **sumar valores poco a poco** dentro de un ciclo.

- Se **inicializa en 0** y, en cada repetici√≥n del ciclo, **se le suma el nuevo valor**.  
- Su funci√≥n principal es **guardar un resultado que se construye paso a paso**.


```python
# Acumulador: suma de los n√∫meros del 1 al 5

numero = 1
suma = 0   # este es el acumulador

# El ciclo se repetir√° mientras 'numero' sea menor o igual a 5
while numero <= 5:
    suma = suma + numero   # el acumulador va guardando la suma total
    numero = numero + 1    # actualizamos la variable de control (evita ciclos infinitos)

print("La suma total es:", suma)


# üìù Ejercicio 2: Mostrar n√∫meros pares del 2 al 20 con `while`

## Enunciado
Escribe un programa en Python que muestre en pantalla todos los n√∫meros **pares** del 2 al 20, utilizando un ciclo `while`.  

Este ejercicio ayuda a:
- Comprender c√≥mo trabajar con incrementos diferentes de 1.
- Practicar condiciones y secuencias espec√≠ficas.


````python
# Ejercicio 3: N√∫meros pares del 2 al 20

# Empezamos desde el primer n√∫mero par
numero = 2

# Mientras el n√∫mero sea menor o igual a 20, se ejecuta el ciclo
while numero <= 20:
    # Mostramos el n√∫mero actual
    print("N√∫mero par:", numero)

    # Como son pares, aumentamos de 2 en 2
    numero = numero + 2

# Cuando termina, ya se han mostrado todos los pares hasta 20
print("Fin del listado de n√∫meros pares ‚úÖ")
````

# üìù Ejercicio 3: Suma de n√∫meros del 1 al 5 con `while`

## Enunciado
Escribe un programa en Python que calcule la suma de los n√∫meros del 1 al 5 utilizando un ciclo `while`.  
El programa debe mostrar en pantalla cada n√∫mero y la suma acumulada paso a paso.  

Este ejercicio permite reforzar:
- El uso de acumuladores.
- La importancia de actualizar correctamente la variable de control.
- C√≥mo visualizar resultados intermedios.


````python
# Ejercicio 2: Suma de los n√∫meros del 1 al 5 con while

# Inicializamos la variable de control
numero = 1

# Inicializamos el acumulador en 0
suma = 0

# Mientras el n√∫mero sea menor o igual a 5, seguimos sumando
while numero <= 5:
    # Mostramos el n√∫mero actual
    print("N√∫mero actual:", numero)

    # Sumamos el n√∫mero al acumulador
    suma = suma + numero

    # Mostramos la suma parcial
    print("Suma acumulada:", suma)

    # Actualizamos la variable de control
    numero = numero + 1

# Mostramos el resultado final
print("La suma total es:", suma)
````

## üîπ 1. La funci√≥n `len()`

- La funci√≥n **`len()`** devuelve la **longitud** (cantidad de elementos) de una cadena, lista, tupla u otro objeto iterable.  
- Es muy usada para recorrer textos o listas con ciclos como `while` o `for`.

üëâ **Ejemplos:**

```python

# Con listas

print(len(numeros))  # Resultado: 4
```

## üîπ 2. El operador `in`

El operador **`in`** se utiliza para **verificar si un elemento est√° dentro de una cadena, lista o conjunto**.  

üëâ Devuelve un valor booleano:

- `False` ‚Üí si el elemento no est√° presente.  

---

### üìå Ejemplos

```python

# Con listas

print("mango" in frutas) # Resultado: False

## ‚úÖ Resumen

- **`len()`** ‚Üí devuelve la cantidad de elementos de una cadena, lista u objeto iterable.
- **`in`** ‚Üí verifica si un elemento est√° presente en una cadena, lista o conjunto.

## Enunciado
Un **n√∫mero primo** es aquel que **solo tiene dos divisores: 1 y √©l mismo**.  
Esto significa que **no puede dividirse exactamente entre ning√∫n otro n√∫mero**.  

üëâ Ejemplos:  
- **Primos:** 2, 3, 5, 7, 11, 13...  
- **No primos:** 4 (divisible entre 1, 2 y 4), 6 (divisible entre 1, 2, 3 y 6).  

Escribe un programa en Python que recorra todos los n√∫meros desde **1 hasta el n√∫mero ingresado** y cuente cu√°ntos divisores tiene.  
- Si tiene exactamente **2 divisores**, es primo.  
- Si tiene m√°s de 2, **no es primo**.  

---

````python
# Ejercicio 11: Identificar si un n√∫mero es primo recorriendo divisores

numero = int(input("Ingresa un n√∫mero entero: "))
divisor = 1
cantidad_divisores = 0

while divisor <= numero:
    if numero % divisor == 0:   # Si el residuo es 0, es divisor exacto
        cantidad_divisores = cantidad_divisores + 1
    divisor = divisor + 1

# Verificamos si el n√∫mero tiene exactamente 2 divisores
if cantidad_divisores == 2:
    print("El n√∫mero", numero, "es primo ‚úÖ")
else:
    print("El n√∫mero", numero, "NO es primo ‚ùå")
````

## Enunciado
Un **n√∫mero primo** es aquel que **solo tiene dos divisores: 1 y √©l mismo**.  

En este ejercicio debes escribir un programa en Python que:  
1. Pida al usuario un n√∫mero entero mayor a 2.

3. Muestre en pantalla cada n√∫mero primo encontrado.  
4. Calcule la **suma total** de esos n√∫meros primos.  

üëâ En esta versi√≥n **no se utilizar√°n listas**, solo variables simples.

---

## üéØ Objetivo
Este ejercicio permite comprender:
- C√≥mo recorrer un rango de n√∫meros con un ciclo `while`.  
- C√≥mo verificar si un n√∫mero es primo mediante divisores.  
- C√≥mo acumular resultados en una variable (suma total).  
- C√≥mo resolver un problema sin usar estructuras de datos avanzadas (listas).  

---

## üí° Reto adicional
1. Modifica el programa para que tambi√©n muestre la **cantidad de n√∫meros primos encontrados**.  
2. Optimiza el algoritmo verificando divisores solo hasta la **ra√≠z cuadrada** del n√∫mero.  
3. Haz que el programa funcione con un **l√≠mite inferior y superior** dados por el usuario (por ejemplo: primos entre 10 y 50).  


````python
# Ejercicio 13: Primos en un rango y suma total (sin listas)

# Pedimos un valor l√≠mite
limite = int(input("Ingresa un n√∫mero mayor a 2: "))

# Inicializamos variables
numero = 2
suma_primos = 0

# Recorremos todos los n√∫meros desde 2 hasta el l√≠mite
while numero <= limite:
    divisor = 1
    cantidad_divisores = 0

    # Verificamos si el n√∫mero es primo contando sus divisores
    while divisor <= numero:
        if numero % divisor == 0:
            cantidad_divisores = cantidad_divisores + 1
        divisor = divisor + 1

    # Si tiene exactamente 2 divisores ‚Üí es primo
    if cantidad_divisores == 2:
        print("Primo encontrado:", numero)
        suma_primos = suma_primos + numero

    numero = numero + 1

# Mostramos la suma total de primos
print("La suma de los primos es:", suma_primos)
````

# üßÆ Ejercicio 14: Factorial de un n√∫mero (con `while`)

## Enunciado
El **factorial** de un n√∫mero entero `n` (escrito `n!`) es el producto de todos los enteros desde `1` hasta `n`.  
Por ejemplo: `5! = 1 √ó 2 √ó 3 √ó 4 √ó 5 = 120`.  
Reglas especiales:
- `0! = 1` por definici√≥n.
- No existe factorial de n√∫meros negativos.

**¬øQu√© es el factorial?**  
Para un entero $n \ge 0$, el factorial $n!$ es el producto $1 \times 2 \times 3 \times \dots \times n$.  
Por convenio, $0! = 1$. No est√° definido para n√∫meros negativos.

**Objetivo did√°ctico**  
- Practicar la estructura b√°sica de un `while` (inicializaci√≥n ‚Üí condici√≥n ‚Üí actualizaci√≥n).  
- Usar un **contador** (`i`) y un **acumulador** (`factorial`).  
- Explicar cada paso del c√°lculo con **impresiones intermedias** (trazabilidad del proceso).

**C√≥digo Comentado**

```python
# 1) ENTRADA: pedimos un entero n
n = int(input("Ingresa un entero n (n >= 0): "))

# 2) VALIDACI√ìN: repetimos hasta que n sea v√°lido (n >= 0)
#    - Si el usuario ingresa un negativo, recordamos la regla y volvemos a pedir el dato.
while n < 0:
    print("El factorial no est√° definido para negativos. Intenta de nuevo.")
    n = int(input("Ingresa un entero n (n >= 0): "))

# 3) INICIALIZACI√ìN DE VARIABLES
#    - 'factorial' es el ACUMULADOR del producto. Inicia en 1 (elemento neutro de la multiplicaci√≥n).
#    - 'i' es el CONTADOR que recorre los valores 1, 2, ..., n.
factorial = 1
i = 1

# 4) TRAZADO INICIAL: informamos que mostraremos cada paso del c√°lculo
print(f"\nC√°lculo de {n}! paso a paso:")

# 5) CICLO PRINCIPAL: mientras i sea menor o igual que n, seguimos multiplicando
#    Patr√≥n del while:
#       a) Usar el valor actual (imprimir/operar)
#       b) Actualizar el acumulador
#       c) Incrementar el contador para acercarnos al fin del ciclo
while i <= n:
    # Guardamos el valor anterior del acumulador para explicar el paso (solo con fines did√°cticos)
    anterior = factorial

    # Actualizamos el acumulador multiplicando por el contador actual
    factorial = factorial * i

    # Explicamos el paso actual con un mensaje claro
    # Ej.: Paso 3: 2 √ó 3 = 6
    print(f"Paso {i}: {anterior} √ó {i} = {factorial}")

    # Actualizamos el contador (¬°evita ciclos infinitos!)
    i = i + 1

# 6) SALIDA: mostramos el resultado final formateado
print(f"\nResultado: {n}! = {factorial}")


````python
# Ejercicio 14: Factorial de un n√∫mero con while (impresi√≥n paso a paso)

n = int(input("Ingresa un entero n (n >= 0): "))

# Validaci√≥n simple
while n < 0:
    print("El factorial no est√° definido para negativos. Intenta de nuevo.")
    n = int(input("Ingresa un entero n (n >= 0): "))

factorial = 1
i = 1

print(f"\nC√°lculo de {n}! paso a paso:")
while i <= n:
    anterior = factorial
    factorial = factorial * i
    print(f"Paso {i}: {anterior} √ó {i} = {factorial}")
    i = i + 1

print(f"\nResultado: {n}! = {factorial}")
````

## Enunciado
Multiplica dos enteros `a` y `b` usando **solo sumas** dentro de un ciclo `while`.  
Ejemplo: `3 √ó 4 = 3 + 3 + 3 + 3`.

**Requisitos:**
- No usar el operador `*` ni funciones que multipliquen.
- Usar √∫nicamente **sumas** (y asignaciones) dentro de un **`while`**.
- Debe funcionar con **valores negativos y cero**.
- Debe **mostrar el proceso paso a paso** (cada suma realizada).

**Pista (opcional):**  
Determina el **signo** del resultado seg√∫n los signos de `a` y `b`. Trabaja con los valores absolutos para contar sumas y **aplica el signo al final**.

## üéØ Objetivo

---

## üí° Reto adicional
1. **Optimiza**: repite la suma el **menor** de `|a|` y `|b|`, sumando el **mayor** para reducir iteraciones.  
2. **Contador de pasos**: al final muestra cu√°ntas **sumas** se realizaron.  
3. **Modo detallado**: permite activar/desactivar la impresi√≥n del paso a paso (√∫til para n√∫meros grandes).  
4. **Validaci√≥n**: pide de nuevo el dato si `a` o `b` no son enteros.


````python

#Ejercicio 15: Multiplicaci√≥n por sumas sucesivas con while (soporta negativos y muestra pasos)

a = int(input("Ingresa a: "))
b = int(input("Ingresa b: "))

# Determinar el signo del resultado usando XOR l√≥gico de negatividad
signo = -1 if (a < 0) ^ (b < 0) else 1

# Trabajamos con valores absolutos para realizar solo sumas positivas
A = abs(a)
B = abs(b)

resultado = 0
contador = 0

print(f"\nC√°lculo de {a} √ó {b} por sumas sucesivas:")
while contador < B:
    anterior = resultado
    resultado = resultado + A
    contador = contador + 1
    print(f"Paso {contador}: {anterior} + {A} = {resultado}")

# Aplicamos el signo al resultado final
resultado = resultado * signo

print(f"\nResultado: {a} √ó {b} = {resultado}")
````

## Enunciado
Escribe un programa en Python que muestre las tablas de multiplicar del **1 al 5**.  

El programa debe usar **dos ciclos `while`**:  
- El primero recorre los n√∫meros base (del 1 al 5).  
- El segundo genera la tabla de cada n√∫mero (del 1 al 10).  

---

## üéØ Objetivo
Este ejercicio permite comprender:
- C√≥mo usar **bucles anidados** (`while` dentro de otro `while`).  
- C√≥mo recorrer distintos rangos con ciclos.  
- C√≥mo generar autom√°ticamente las **tablas de multiplicar** en forma ordenada.  

---

## üí° Reto adicional

````python
# Ejercicio 16: Tablas de multiplicar con dos while

base = 1

while base <= 5:   # Primer while: controla qu√© tabla se muestra
    print(f"\nTabla del {base}:")

    multiplicador = 1
    while multiplicador <= 10:   # Segundo while: genera cada tabla
        resultado = base * multiplicador
        print(f"{base} x {multiplicador} = {resultado}")
        multiplicador = multiplicador + 1

    base = base + 1
````

## Enunciado
Haz un programa que dibuje una **pir√°mide de asteriscos** de altura `n` pedida al usuario.  

Debe usar **dos ciclos `while`**:  
- El primero controla el n√∫mero de filas.  
- El segundo imprime los asteriscos de cada fila.  

---

## üéØ Objetivo
Este ejercicio permite comprender:
- C√≥mo trabajar con **bucles anidados** en Python.  
- C√≥mo generar patrones visuales con caracteres.  
- C√≥mo controlar la cantidad de repeticiones en funci√≥n de una entrada del usuario.  

---

## üí° Reto adicional
1. Haz que el programa pueda generar tambi√©n una **pir√°mide invertida**.  
2. Permite al usuario elegir el **s√≠mbolo** de la pir√°mide (por ejemplo, `#`, `+`, `@`).  



````python
# Ejercicio 17: Pir√°mide de asteriscos con dos while

n = int(input("Ingresa la altura de la pir√°mide: "))

fila = 1
while fila <= n:   # Primer while: controla las filas
    col = 1
    while col <= fila:   # Segundo while: imprime asteriscos en la fila
        print("*", end="")  # end="" evita el salto de l√≠nea inmediato
        col = col + 1
    print()  # salto de l√≠nea despu√©s de terminar la fila
    fila = fila + 1
````
